```python
new_cols_dict = {
'timestamp' : 'time',
'cnt' : 'count',
't1' : 'temp_real_C ',
't2' : 'temp_feels_like_C ',
'hum' : 'humidity_percent',
'windspeed' : 'windspeed_kph',
'weather_code' : 'weather',
'is_holiday' : 'is_holiday',
'is_weekend' : 'is_weekend',
'season' : 'season'
   
}
# Making the changes in columns permanent by using inplace  = True
bikes.rename(new_cols_dict, axis=1, inplace=True)
```

```python
bikes.head()
```

```python
# Changing the humidity values between 0 and 1 
bikes.humidity_percent = bikes.humidity_percent/100
```

```python
bikes.season.describe()
```

```python
# Creating the season directory so that we can actually map the integers 0-3 to written values 

season_dict = {
'0.0' : 'Spring',
'1.0' : 'Summer',
'2.0' : 'Autumn',
'3.0' : 'Winter'
} 

# Creating the weather directory so that we can actually map the integers to written values 

weather_dict = {
'1.0' : 'Clear', 
'2.0' : 'Scattered Cloud',
'3.0' : 'Broken Clouds',
'4.0' : 'Cloudy',
'7.0' : 'Rain',
'10.0' : 'Rain with thunderstorm',
'26.0' : 'Snowfall'
}

# Changing the season data type to string 
bikes.season = bikes.season.astype('str')

# Mapping the values 0-3 to actual written seasons 

bikes.season = bikes.season.map(season_dict)


# Changing the weather data type to string 
bikes.weather = bikes.weather.astype('str')

# Mapping the values 0-3 to actual written seasons 

bikes.weather = bikes.weather.map(weather_dict)
```
